#2 Example with code
##2.1 Import Libraries


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal, ndimage
from scipy.misc import ascent  # Deprecated, use scipy.datasets.ascent instead
```

##2.2 Signal Processing
###2.2.1 Generate a random signal
In this section, a random signal is generated using the np.random.normal function to represent noise.

```python
# Generate a random signal
np.random.seed(0)
t = np.linspace(0, 1, 500)
signal_noise = np.random.normal(loc=0, scale=0.5, size=len(t))
signal_clean = np.sin(2 * np.pi * 10 * t) + np.sin(2 * np.pi * 20 * t)
signal_noisy = signal_clean + signal_noise
```

###2.2.2 Apply a low-pass filter
A low-pass Butterworth filter is applied to the noisy signal using the signal.butter and signal.filtfilt functions from SciPy's signal module. This helps remove high-frequency noise from the signal.

```python
# Apply a low-pass filter to remove high-frequency noise
cutoff_freq = 0.2  # Set the cutoff frequency (less than fs/2)
b, a = signal.butter(4, cutoff_freq, fs=2, btype='low')
filtered_signal = signal.filtfilt(b, a, signal_noisy)
```

###2.2.3 Perform a Fourier transform
 The signal.periodogram function is used to compute the frequency spectrum of the noisy signal. This allows us to analyze the frequency content of the signal.

```python
# Perform a Fourier transform to analyze the frequency content of the signal
freq, spectrum = signal.periodogram(signal_noisy, fs=2)
```

###2.2.4 Plot the signals:
The original signal, noisy signal, and filtered signal are plotted using plt.plot from Matplotlib. The clean signal is also included for comparison. The frequency spectrum is plotted using plt.plot.

```python
# Plot the original signal, noisy signal, and filtered signal
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.plot(t, signal_clean, label='Clean Signal')
plt.plot(t, signal_noisy, label='Noisy Signal')
plt.plot(t, filtered_signal, label='Filtered Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()

# Plot the frequency spectrum
plt.subplot(2, 1, 2)
plt.plot(freq, spectrum)
plt.xlabel('Frequency')
plt.ylabel('Power Spectrum')
plt.show()
```

##2.3 Image Manipulation
###2.3.1 Read and display an image:
The ascent function from scipy.misc (deprecated) is used to load the "ascent" image, which is a sample image provided by SciPy. It is displayed using plt.imshow with the grayscale colormap.

```python
# Read and display an image
image = ascent()  # Deprecated, use scipy.datasets.ascent() instead
plt.figure()
plt.imshow(image, cmap='gray')
plt.axis('off')
plt.title('Original Image')
```

###2.3.2 Apply a Gaussian filter
The ndimage.gaussian_filter function is used to apply a Gaussian filter to the image, resulting in a blurred version of the image.

```python
# Apply a Gaussian filter to blur the image
blurred_image = ndimage.gaussian_filter(image, sigma=2)
```

###2.3.3 Resize the image:
The ndimage.zoom function is used to resize the image to a smaller size. In this case, the image is resized to half its original dimensions.

```python
# Resize the image to a smaller size
resized_image = ndimage.zoom(image, 0.5)
```

###2.3.4 Rotate the image
 The ndimage.rotate function is used to rotate the image by a specified angle. In this example, the image is rotated by 45 degrees.

```python
# Rotate the image
rotated_image = ndimage.rotate(image, angle=45)
```

###2.3.5 Flip the image vertically
The np.flipud function is used to flip the image vertically, effectively reversing the order of rows in the image.

```python
# Flip the image vertically
flipped_image = np.flipud(image)
```

###2.3.6 Plot the manipulated images:
The blurred image, resized image, rotated image, and flipped image are plotted using plt.imshow with the grayscale colormap. The images are displayed in a grid layout using plt.subplot.

```python
# Plot the manipulated images
plt.figure(figsize=(10, 6))
plt.subplot(2, 2, 1)
plt.imshow(blurred_image, cmap='gray')
plt.axis('off')
plt.title('Blurred Image')

plt.subplot(2, 2, 2)
plt.imshow(resized_image, cmap='gray')
plt.axis('off')
plt.title('Resized Image')

plt.subplot(2, 2, 3)
plt.imshow(rotated_image, cmap='gray')
plt.axis('off')
plt.title('Rotated Image')

plt.subplot(2, 2, 4)
plt.imshow(flipped_image, cmap='gray')
plt.axis('off')
plt.title('Flipped Image')

plt.tight_layout()
plt.show()
```
